```python
import matplotlib.pyplot as plt

# Dados
anos = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
adesao_manual = [55, 60, 65, 70, 72, 75, 76, 77]
adesao_automatica = [0, 0, 75, 80, 85, 90, 94, 96]
permanencia = [0, 0, 80, 85, 90, 92, 95, 96]

# Criar figura e eixo
fig, ax = plt.subplots()

# Plotar gráfico de barra
ax.bar(anos, adesao_manual, label='Adesão Manual')
ax.bar(anos, adesao_automatica, label='Adesão Automática')
ax.bar(anos, permanencia, label='Permanência')

# Configurar título e legendas
ax.set_title('Taxa de Adesão e Permanência')
ax.set_xlabel('Ano')
ax.set_ylabel('Taxa de Adesão (%)')
ax.legend()

# Mostrar gráfico
plt.show()
```
