```python
# Primeiro, instale a biblioteca graphviz se estiver usando o Google Colab
!pip install graphviz

# Importar a biblioteca
from graphviz import Digraph

# Criar um novo gráfico
dot = Digraph()

# Definindo as entidades
dot.node('D', 'Departamento\n(ID_Departamento)\nNome\nSala\nAndar\nBloco', shape='box')
dot.node('P', 'Projeto\n(ID_Projeto)\nNome\nDescrição\nHoras_Previstas\nData_Início\nData_Término\nID_Departamento', shape='box')
dot.node('C', 'Colaborador\n(ID_Colaborador)\nNome\nData_Nascimento\nCPF\nSexo\nEstado_Civil\nTelefone\nEmail', shape='box')
dot.node('A', 'Alocação\n(ID_Alocação)\nID_Colaborador\nID_Projeto\nHoras_Alocação', shape='box')

# Definindo os relacionamentos
dot.edge('D', 'P', label='1:N')
dot.edge('C', 'A', label='N')
dot.edge('P', 'A', label='N')
dot.edge('D', 'P', label='N:1')

# Renderizar o gráfico
dot.render('modelo_der', format='png', cleanup=True)

# Exibir o gráfico
dot
```
