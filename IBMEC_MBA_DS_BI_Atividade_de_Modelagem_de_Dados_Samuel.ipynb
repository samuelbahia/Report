```python
# Primeiro, instale a biblioteca graphviz se estiver usando o Google Colab
!pip install graphviz

# Importar a biblioteca
from graphviz import Digraph

# Criar um novo gráfico
dot = Digraph(format='png', engine='dot')

# Definindo as entidades com um tamanho de nó adequado
dot.node('P', 'Produto\n(ID_Produto)\nNome\nPreço\nDescrição', shape='box', width='2', height='1', fontsize='10')
dot.node('F', 'Fornecedor\n(ID_Fornecedor)\nNome\nEndereço\nTelefone\nEmail', shape='box', width='2', height='1', fontsize='10')
dot.node('C', 'Cliente\n(ID_Cliente)\nNome\nEndereço\nTelefone\nEmail\nClassificação', shape='box', width='2', height='1', fontsize='10')
dot.node('CP', 'Compra\n(ID_Compra)\nData_Hora\nQuantidade\nID_Cliente\nID_Produto', shape='box', width='2', height='1', fontsize='10')

# Definindo os relacionamentos
dot.edge('P', 'F', label='N:M')
dot.edge('C', 'CP', label='1:N')
dot.edge('P', 'CP', label='N:M')

# Renderizar o gráfico
dot.render('modelo_der_loja', cleanup=True)

# Exibir o gráfico
dot
```
